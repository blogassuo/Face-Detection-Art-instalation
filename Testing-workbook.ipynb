```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Load the classifier model
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# Capture the frame from the webcam
def capture():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    ret, frame = cap.read()
    return frame

#Display image using matplotlib
def display(frame):
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))
    plt.title('Captured frame')
    plt.show()
```

```python
frame = capture()
frame.shape
```

```python
# Release the camera
cap.release()
```

```python
def detect(frame, minsize):
    faces = face_cascade.detectMultiScale(frame, scaleFactor=1.3, minNeighbors=6, minSize=(minsize, minsize))
    return faces

def draw_rec(frame, detected_faces):
    if len(detected_faces) > 0:
        for (x,y,w,h) in detected_faces:
            cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
        
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))
        plt.show()
```
